Acerca del conjunto de datos<br>
**Comercialización bancaria**<br>
*Resumen:*<br>
Los datos están relacionados con campañas de marketing directo (llamadas telefónicas) de una institución bancaria portuguesa. El objetivo de la clasificación es predecir si el cliente suscribirá un depósito a plazo (variable y).<br>

*Información del conjunto de datos:*<br>
Los datos están relacionados con campañas de marketing directo de una institución bancaria portuguesa. Las campañas de marketing se basaron en llamadas telefónicas. Muchas veces era necesario más de un contacto con un mismo cliente, para poder acceder si el producto (depósito a plazo bancario) estaría ('sí') o no ('no') suscrito.<br>

*Información de atributos:*<br>
**Datos del cliente bancario:**<br>
1. Edad (numérica)<br>
2. **Trabajo** : tipo de trabajo (categórico: 'administrador', 'obrero', 'emprendedor', 'criada', 'administración', 'jubilado', 'autónomo', 'servicios', 'estudiante', 'técnico', 'desempleado', 'desconocido')<br>
3. **Civil**: estado civil (categórico: 'divorciado', 'casado', 'soltero', 'desconocido'; nota: 'divorciado' significa divorciado o viudo)<br>
4. **Educación** (categórica: 'básico.4y', 'básico.6y', 'básico.9y', 'bachillerato', 'analfabeto', 'curso.profesional', 'título.universitario', 'desconocido')<br>
5. **Incumplimiento**: ¿tiene crédito en incumplimiento? (categórico: 'no', 'sí', 'desconocido')<br>
6. **Vivienda**: ¿tiene préstamo de vivienda? (categórico: 'no', 'sí', 'desconocido')<br>
7. **Préstamo**: ¿tiene préstamo personal? (categórico: 'no', 'sí', 'desconocido')<br>

**Relacionado con el último contacto de la campaña actual:**<br>
8. **Contacto**: tipo de comunicación del contacto (categórico: 'celular', 'teléfono')<br>
9. **Mes**: mes del año del último contacto (categórico: 'ene', 'feb', 'mar', ..., 'nov', 'dic')<br>
10. **Day_of_week**: último día de contacto de la semana (categórico: 'lunes', 'martes', 'miércoles', 'jueves', 'viernes')<br>
11. **Duración**: duración del último contacto, en segundos (numérico). Nota importante: este atributo afecta en gran medida el objetivo de salida (por ejemplo, si duración=0 entonces y='no'). Sin embargo, no se conoce la duración antes de realizar una llamada. Además, después del final de la llamada, obviamente se conoce y.<br>

**Otros atributos:**<br>
12. **Campaña**: número de contactos realizados durante esta campaña y para este cliente (numérico, incluye el último contacto)<br>
13. **Pdays**: número de días que pasaron después de que el cliente fue contactado por última vez desde una campaña anterior (numérico; 999 significa que el cliente no fue contactado previamente)<br>
14. **Anterior**: número de contactos realizados antes de esta campaña y para este cliente (numérico)<br>
15. **Resultado**: resultado de la campaña de marketing anterior (categórico: 'fracaso', 'inexistente', 'éxito')<br>

**Atributos del contexto social y económico:**<br>
16. **Emp.var.rate**: tasa de variación del empleo - indicador trimestral (numérico)<br>
17. **Cons.price.idx**: índice de precios al consumidor - indicador mensual (numérico)<br>
18. **Cons.conf.idx**: índice de confianza del consumidor - indicador mensual (numérico)<br>
19. **Euribor3m**: tipo de cambio euribor a 3 meses - indicador diario (numérico)<br>
20. **Nr.empleados**: número de empleados - indicador trimestral (numérico)<br>

**Variable de salida (objetivo deseado):**<br>
**y** - ¿el cliente ha suscrito un depósito a plazo? (binario: 'sí', 'no')<br>

*Pasos de análisis:*<br>
1. Análisis de información de atributos.
2. Aprendizaje automático (regresión logística, KNN, SVM, árbol de decisión, bosque aleatorio, Naive Bayes)
3. Aprendizaje profundo (ANN)

# Data Transform
Lo primero que vamos a hacer es eliminar columnas que no usarmos. Luego, vamos a convertir todas nuestras columnas a tipo numericas, por ultimo, le aplicaremos one hot encoder.

```python
#eliminamos la variable duration
del bank["duration"] #usamos este metodo para que se elimine directamente del df
```

```python
#concatenamos las variables month y day_of_week, luego la pasamos a formato fecha y por ultimo borramos el año porque no lo tenemos y por defecto va a ser 1990
variables_fecha = ["month","day_of_week"]
bank['fecha'] = bank['month'] + ' ' + bank['day_of_week']
```

```python
bank['fecha'] = pd.to_datetime(bank['fecha'], format='%b %a')
bank.head(10)
```

```python
bank = bank.drop(['month', 'day_of_week'], axis=1)
```

**PDAYS**: Vamos a crear una variable nueva para obtener si el cliente fue contactado antes o no, ya que tenemos que si indica 999 es porque no se contacto.

```python
bank['llamada_cliente'] = (bank['pdays'] != 999).astype(int)
bank['dias_desde_ultimo_contacto'] = bank['pdays'].replace(999, 0)
```

```python
print(bank["llamada_cliente"].value_counts())
print("---------------------")
print(bank["pdays"].value_counts().index)
print("---------------------")
print(bank["dias_desde_ultimo_contacto"].value_counts().index)
```
